```python
from pykrx import stock
import pandas as pd

# 파일 경로 설정
file_path = r'C:\PythonProjects\recon\positions.xlsx'

# 'start_position' 시트의 데이터를 불러오기
df = pd.read_excel(file_path, sheet_name='start_position', dtype={'ticker': str})

# ticker 열의 종목들에 대한 종가 데이터 불러오기
closing_prices = []
for ticker in df['ticker']:
    try:
        # pykrx의 get_market_ohlcv_by_date 함수를 사용하여 종가 데이터 불러오기
        close_price = stock.get_market_ohlcv_by_date("20240101", "20240310", ticker)["종가"][-1]
        closing_prices.append(close_price)
    except Exception as e:
        print(f"Error fetching data for ticker {ticker}: {e}")
        closing_prices.append(None)  # 에러가 발생하면 None으로 처리

# 종가 데이터를 데이터프레임에 추가
df['closing_price'] = closing_prices

# 결과를 'start' 시트에 저장
with pd.ExcelWriter(file_path, mode='a', engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='start', index=False)

# 결과 출력
print("DataFrame이 'start' 시트에 저장되었습니다.")
```

```python
# HINTs #22713에서 받은 기간내 거래내역 불러오기 및 result_df.xlsx와 정보 매칭 작업
# 해당 작업은 oms 및 trade팀의 부정확한 단가 및 거래금액, 수수료, 거래세를 교정하기 위한 작업

# 필요한 라이브러리 불러오기
import pandas as pd
import datetime
from openpyxl import load_workbook

# 당일 날짜 저장
now = datetime.datetime.today().strftime('%Y-%m-%d')

# 거래내역 위치 저장
file_path_df1 = "C:\\PythonProjects\\recon\\hints\\2024-03-19.xlsx"
file_path_df2 = r'C:\PythonProjects\recon\result_df.xlsx'

# 당일 거래내역 데이터프레임으로 저장
df1 = pd.read_excel(file_path_df1, header=0)
df2 = pd.read_excel(file_path_df2, header=0)

# '종목명' 열이 'NaN'인 행 삭제
df1 = df1.dropna(subset=['종목명'])

# '날짜'인덱스의 이름이 일치하지 않으므로 '일자' 인덱스 이름을 변경
df1 = df1.rename(columns={'일자':'날짜'})
df2 = df2.rename(columns={'펀드':'펀드코드'})
df2 = df2.rename(columns={'체결수량':'수량'})

# '매매구분' 칼럼 값 변경
df1.loc[df1['매매구분'] == '매수', '매매구분'] = 'Buy'
df1.loc[df1['매매구분'] == '매도', '매매구분'] = 'Sell'

# '날짜'컬럼의 데이터형식 문제로 인해 통합되지 않는 문제 해결
df1['날짜'] = pd.to_datetime(df1['날짜'], format='%Y-%m-%d')
df2['날짜'] = pd.to_datetime(df2['날짜'], format='%Y-%m-%d')

# 'df1'에 단축코드 컬럼 추가
# df1['단축코드'] = df1['종목코드'].str[3:9]


# 날짜, 종목명, 펀드코드, 매매구분
merged_df = pd.merge(df1, df2, on=['날짜', '종목명', '펀드코드', '매매구분'], how='inner')

# 중복되는 행 삭제
merged_df_no_duple = merged_df.drop_duplicates()

# 경로 및 시트이름 설정
output_file = r'C:\PythonProjects\recon\positions.xlsx'
sheet_name = 'transaction'

# 결과를 엑셀 파일로 저장합니다.
try:
    # 기존 엑셀 파일 열기
    wb = load_workbook(output_file)
    writer = pd.ExcelWriter(output_file, engine='openpyxl')
    writer.book = wb
except FileNotFoundError:
    # 파일이 없을 경우 새로운 엑셀 파일 생성
    writer = pd.ExcelWriter(output_file, engine='openpyxl')

merged_df_no_duple.to_excel(writer, sheet_name=sheet_name, index=False)
writer.save()
writer.close()
```

```python
import pandas as pd

# 엑셀 파일에서 시트 데이터를 읽어옵니다.
realized_pnl_df = pd.read_excel('your_excel_file.xlsx', sheet_name='realized_pnl')
unrealized_pnl_df = pd.read_excel('your_excel_file.xlsx', sheet_name='unrealized_pnl')
start_df = pd.read_excel('your_excel_file.xlsx', sheet_name='start')
transaction_df = pd.read_excel('your_excel_file.xlsx', sheet_name='transaction')

# 필요한 열만 선택하여 데이터프레임을 생성합니다.
realized_pnl_df = realized_pnl_df[['fund_code', 'ticker', 'PnL']]
unrealized_pnl_df = unrealized_pnl_df[['fund_code', 'ticker', 'unrealized_pnl']]
start_df = start_df[['fund_code', 'ticker', 'quantity', 'price']]
transaction_df = transaction_df[['fund_code', 'ticker', 'direction', 'quantity', 'price', 'net_amount']]

# 네번째 시트의 거래 데이터를 기존 포지션에 반영하여 첫번째 시트의 realized_pnl을 계산합니다.
for index, transaction in transaction_df.iterrows():
    fund_code = transaction['fund_code']
    ticker = transaction['ticker']
    direction = transaction['direction']
    quantity = transaction['quantity']
    price = transaction['price']
    net_amount = transaction['net_amount']
    
    # 매수인 경우
    if direction == '매수':
        # 기존 포지션에 추가
        existing_position = start_df[(start_df['fund_code'] == fund_code) & (start_df['ticker'] == ticker)]
        if not existing_position.empty:
            start_df.loc[(start_df['fund_code'] == fund_code) & (start_df['ticker'] == ticker), 'quantity'] += quantity
        else:
            # 새로운 포지션 추가
            start_df = start_df.append({'fund_code': fund_code, 'ticker': ticker, 'quantity': quantity, 'price': price}, ignore_index=True)
    # 매도인 경우
    elif direction == '매도':
        # 기존 포지션에서 빼기
        existing_position = start_df[(start_df['fund_code'] == fund_code) & (start_df['ticker'] == ticker)]
        if not existing_position.empty:
            start_df.loc[(start_df['fund_code'] == fund_code) & (start_df['ticker'] == ticker), 'quantity'] -= quantity

# realized_pnl을 계산하여 첫번째 시트에 채웁니다.
for index, row in realized_pnl_df.iterrows():
    fund_code = row['fund_code']
    ticker = row['ticker']
    quantity = start_df[(start_df['fund_code'] == fund_code) & (start_df['ticker'] == ticker)]['quantity'].values[0]
    price = start_df[(start_df['fund_code'] == fund_code) & (start_df['ticker'] == ticker)]['price'].values[0]
    realized_pnl_df.loc[index, 'PnL'] = quantity * (price - row['PnL'])

# unrealized_pnl을 계산하여 두번째 시트에 채웁니다.
for index, row in unrealized_pnl_df.iterrows():
    fund_code = row['fund_code']
    ticker = row['ticker']
    quantity = start_df[(start_df['fund_code'] == fund_code) & (start_df['ticker'] == ticker)]['quantity'].values[0]
    price = start_df[(start_df['fund_code'] == fund_code) & (start_df['ticker'] == ticker)]['price'].values[0]
    unrealized_pnl_df.loc[index, 'unrealized_pnl'] = quantity * (price - row['unrealized_pnl'])

# 결과를 엑셀 파일로 저장합니다.
with pd.ExcelWriter('output.xlsx') as writer:
    realized_pnl_df.to_excel(writer, sheet_name='realized_pnl', index=False)
    unrealized_pnl_df.to_excel(writer, sheet_name='unrealized_pnl', index=False)
```

```python
import pandas as pd

# 파일 경로
file_path = r'C:\PythonProjects\recon\positions.xlsx'


# 엑셀 파일에서 시트 데이터를 읽어옵니다.
start_df = pd.read_excel(file_path, sheet_name='start')
transaction_df = pd.read_excel(file_path, sheet_name='transaction')

# Modify start_df based on direction
start_df["quantity"] = start_df.apply(lambda row: f"+{row['quantity']}" if row['direction'] == "LONG" else f"-{row['quantity']}", axis=1)
start_df["net_amount"] = start_df.apply(lambda row: f"+{row['net_amount']}", axis=1)

# 4번째 열(column) 선택 후 문자열로 변환하여 저장 Select column 4 and change data to string type
transaction_df.iloc[:, 2] = transaction_df.iloc[:, 2].apply(lambda x: str(x).zfill(6))
start_df.iloc[:, 1] = start_df.iloc[:, 1].apply(lambda x: str(x).zfill(6))

# Modify transaction_df based on direction, tax, and commission
transaction_df["quantity"] = transaction_df.apply(lambda row: f"+{row['quantity']}" if row['direction'] == "매수" else f"-{row['quantity']}", axis=1)
transaction_df["gross_amount"] = transaction_df.apply(lambda row: f"+{row['gross_amount']}" if row['direction'] == "매수" else f"-{row['gross_amount']}", axis=1)
transaction_df["net_amount"] = transaction_df.apply(lambda row: f"+{row['net_amount']}" if row['direction'] == "매수" else f"-{row['net_amount']}", axis=1)
transaction_df["tax"] = transaction_df.apply(lambda row: f"-{row['tax']}", axis=1)
transaction_df["commission"] = transaction_df.apply(lambda row: f"-{row['commission']}", axis=1)

```

```python
transaction_df[['quantity', 'gross_amount', 'commission', 'tax', 'net_amount', 'price']] = transaction_df[['quantity', 'gross_amount', 'commission', 'tax', 'net_amount', 'price']].astype(float)
```

```python
# transaction_df에서 date 열 제거
transaction_df.drop(columns=['date'], inplace=True)

# fund_code, ticker, manager 열이 같은 행의 값을 합하고 price는 평균값으로 계산하여 하나의 행으로 만듦
result_df = transaction_df.groupby(['fund_code', 'ticker', 'manager']).agg({
    'quantity': 'sum',
    'gross_amount': 'sum',
    'commission': 'sum',
    'tax': 'sum',
    'net_amount': 'sum',
    'price': 'mean'
}).reset_index()

print(result_df)
```
