```python
import pykrx
from pykrx import stock
import pandas as pd
import numpy as np

# 읽어들일 파일 및 시트의 이름 저장.
file_path = r'C:\Users\jtlee\OneDrive - DUNAMIS INVESTMENT ADVISORY\바탕 화면\기타\Maven\작업v1.xlsx'
sheet_name1 = '2020 realized'
sheet_name2 = '2021 realized'
sheet_name3 = '2022 realized'
sheet_name4 = '2023 realized'

# 엑셀파일과 시트 읽기 및 저장
df1 = pd.read_excel(file_path, sheet_name=sheet_name1)
df2 = pd.read_excel(file_path, sheet_name=sheet_name2)
df3 = pd.read_excel(file_path, sheet_name=sheet_name3)
df4 = pd.read_excel(file_path, sheet_name=sheet_name4)

# 티커를 6자리로 지정 및 저장
df1.iloc[:, 1] = df1.iloc[:, 1].apply(lambda x: str(x).zfill(6))
df2.iloc[:, 1] = df2.iloc[:, 1].apply(lambda x: str(x).zfill(6))
df3.iloc[:, 1] = df3.iloc[:, 1].apply(lambda x: str(x).zfill(6))
df4.iloc[:, 1] = df4.iloc[:, 1].apply(lambda x: str(x).zfill(6))

# 데이터를 얻고 싶은 날짜 지정
date1 = "20200630"
date2 = "20210630"
date3 = "20220630"
date4 = "20230630"

# 2020-06-30의 종가를 가져와서 새로운 열에 추가
def get_close_price(ticker, date):
    try:
        close_price = stock.get_market_ohlcv_by_date(date, date, ticker)['종가'].values[0]
        return close_price
    except:
        return np.nan

df1['end_date_close_price'] = df1['ticker'].apply(lambda x: get_close_price(x,date1))
df2['end_date_close_price'] = df2['ticker'].apply(lambda x: get_close_price(x,date2))
df3['end_date_close_price'] = df3['ticker'].apply(lambda x: get_close_price(x,date3))
df4['end_date_close_price'] = df4['ticker'].apply(lambda x: get_close_price(x,date4))

# 결과를 새로운 시트 '2020 realized_v'로 저장
with pd.ExcelWriter(file_path, mode='a', if_sheet_exists='replace') as writer:
    df1.to_excel(writer, sheet_name='2020 realized_v', index=False)
with pd.ExcelWriter(file_path, mode='a', if_sheet_exists='replace') as writer:
    df2.to_excel(writer, sheet_name='2021 realized_v', index=False)
with pd.ExcelWriter(file_path, mode='a', if_sheet_exists='replace') as writer:
    df3.to_excel(writer, sheet_name='2022 realized_v', index=False)
with pd.ExcelWriter(file_path, mode='a', if_sheet_exists='replace') as writer:
    df4.to_excel(writer, sheet_name='2023 realized_v', index=False)

```
