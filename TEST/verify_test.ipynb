```python
import numpy as np
import pandas as pd
from tabulate import tabulate


oms_df = pd.read_excel("C:/PythonProjects/data/20230503_oms.xls")
trader_df = pd.read_excel("C:/PythonProjects/data/20230503_oms.xls")

oms_df1 = oms_df[["펀드", "종목명", "매매유형", "매매구분", "체결수량", "체결단가", "체결금액"]]

oms_df1["매매구분"] = oms_df1[["매매유형", "매매구분"]].agg('_'.join, axis=1)

oms_df1["매매구분"] = oms_df1["매매구분"].replace(['일반_매수', '일반_매도', '차입_매수', '차입_매도'], ['Buy', 'Sell', 'Buy cover', 'Short sell'])

oms_df1.drop(columns=["매매유형"], inplace=True)

oms_df1.dropna(inplace=True)

oms_df1["체결수량"] = oms_df1["체결수량"].astype(int)

oms_df1["체결단가"] = oms_df1["체결단가"].astype(int)

oms_df1["체결금액"] = oms_df1["체결금액"].astype(int)

oms_df1.set_index(['펀드', '종목명', '매매구분'], inplace=True)

oms_df1.sort_index(axis=0, inplace=True)

oms_df1 = oms_df1[(oms_df1 != 0).all(axis=1)]

oms_df1.reset_index(inplace=True)

oms_df1.drop(oms_df1[oms_df1['종목명'].str.contains('KOSPI200F|KOSDAQ150F|코스피200F|코스닥150F|코스피F200|코스닥F150')].index, inplace=True)

trader_df.dropna(inplace=True)

trader_df["체결수량"] = trader_df["체결수량"].astype(int)

trader_df["체결단가"] = trader_df["체결단가"].astype(int)

trader_df["체결금액"] = trader_df["체결금액"].abs().astype(int)

trader_df.set_index(['펀드', '종목명', '매매구분'], inplace=True)

# trader_df.drop(index=trader_df[(trader_df['매매처'] == 'KIS Swap') | (trader_df['매매처'] == 'KB Swap')].index, inplace=True)

# trader_df.drop(columns=["매매처"], inplace=True)

trader_df.sort_index(axis=0, inplace=True)

trader_df.reset_index(inplace=True)

trader_df.drop(trader_df[trader_df['종목명'].str.contains('KOSPI200F|KOSDAQ150F|코스피200F|코스닥150F|코스피F200|코스닥F150')].index, inplace=True)

result_df = oms_df1.merge(trader_df, how='outer', on=['펀드', '종목명', '매매구분'], suffixes=('_oms', '_trader'))

result_df["체결수량_oms"].fillna(0, inplace=True)

result_df["체결단가_oms"].fillna(0, inplace=True)

result_df["체결금액_oms"].fillna(0, inplace=True)

result_df["체결수량_trader"].fillna(0, inplace=True)

result_df["체결단가_trader"].fillna(0, inplace=True)

result_df["체결금액_trader"].fillna(0, inplace=True)

result_df["체결수량"] = result_df["체결수량_oms"] - result_df["체결수량_trader"]

result_df["체결단가"] = result_df["체결단가_oms"] - result_df["체결단가_trader"]

result_df["체결금액"] = result_df["체결금액_oms"] - result_df["체결금액_trader"]
```
